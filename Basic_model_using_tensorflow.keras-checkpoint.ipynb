### Data Preparation and Preprocessing

```python
# Importing required libraries
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Example data:
- An example drug was tested on individuals from ages 13 to 100 in a clinical trial
- The trial had 2100 participants. Half of them were under 65 years old and half were 65 years or older
- Around 95% patients over 65 years experienced side effects
- Around 95% patients under 65 years not experienced any side effects

```python
# Creating dataset
for i in range(50):
    # The ~5% of younger individuals who experienced side effects
    random_younger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(1)  # 1 represents patient experienced side effects
    
    # The ~5% of older individuals who did not experienced side effects
    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(0)  # 0 represents patient did not experienced side effects
    
for i in range(1000):
    # The ~95% of younger individuals who did not experienced any side effects
    random_younger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    # The ~95% of older individuals who experienced side effects
    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
# We must provide inputs to Tensorflow models as numpy array or tensorflow tensor hence converting data to numpy array
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples) # shuffle both labels and sample to avoid any patterns
```

```python
# Scaling the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1)) # reshaping data as fit_transform doesn't accept 1-d data
```

```python
for i in scaled_train_samples:
    print(i)
```

### Building simple tensorflow.keras Sequential model

```python
# Importing required libraries 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'), # first hidden and input layer
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax') # output layer
])
```

```python
model.summary()
```

```python
# Compiling the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training the model
model.fit(x=scaled_train_samples, y=train_labels, batch_size=10, epochs=30, shuffle=True, verbose=2) 
```

```python
# Creating model with Validation set
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
# training set is shuffled after splitting Validation Set so we should make sure that training set is shuffles prior if 
# we're using 'validation_split' function
```

### Cheking model accuracy on test dataset

```python
# Creating the test data
test_labels = []
test_samples = []
```

```python
# Creating dataset
for i in range(10):
    # The ~5% of younger individuals who experienced side effects
    random_younger = randint(13, 64)
    test_samples.append(random_younger)
    test_labels.append(1)  # 1 represents patient experienced side effects
    
    # The ~5% of older individuals who did not experienced side effects
    random_older = randint(65, 100)
    test_samples.append(random_older)
    test_labels.append(0)  # 0 represents patient did not experienced side effects
    
for i in range(200):
    # The ~95% of younger individuals who did not experienced any side effects
    random_younger = randint(13, 64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    # The ~95% of older individuals who experienced side effects
    random_older = randint(65, 100)
    test_samples.append(random_older)
    test_labels.append(1)
```

```python
# We must provide inputs to Tensorflow models as numpy array or tensorflow tensor hence converting data to numpy array
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
test_labels, test_samples = shuffle(test_labels, test_samples) # shuffle both labels and sample to avoid any patterns
```

```python
# Scaling the test datasets
scaler = MinMaxScaler(feature_range=(0,1))
scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

### Making predictions using model

```python
predictions = model.predict(x=scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in predictions:
    print(i)
    
# we're getting probability as predictions
```

```python
for i in rounded_predictions:
    print(i)
```

### Visualizing performance using Confusion Matrix

```python
# Importing required libraries 
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes, 
                          normalize=False, 
                          title='Confusion Matrix', 
                          cmap =plt.cm.Blues):
    # this function prints & plots the confusion matrix 
    # normalization can be applied by setting 'normalize=True'
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print('Normalized Confusion Matrix')
    else:
        print('Confusion Matrix(without normalization)')

    print(cm)
    
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment='center',
                 color='white' if cm[i, j] > thresh else 'black')
        plt.tight_layout()
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
```

```python
cm_plot_labels = ['no_side_effects', 'side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```

### Save and Load Model

#### 1. model.save()

```python
# First check if the file already exists or not
# If not, model is saved to the disk
import os.path
if os.path.isfile('models/medical_trial_model.h5') is False:
    model.save('models/medical_trial_model.h5')
```

The save function saves:
- The architecture of the model, allowing to re-create the model
- The weights of the model
- The training configuration(loss, optimizer)
- The state of optimizer, allowing to resume training exactly where you left off

```python
# Loading saved models
from tensorflow.keras.models import load_model
new_model = load_model('models/medical_trial_model.h5')
```
